# Methodology

### Combine Relevant Features into "job_content"

```
df['job_content'] = (
    df['title'].astype(str) + ' ' +
    df['company_profile'].astype(str) + ' ' +
    df['description'].astype(str) + ' ' +
    df['requirements'].astype(str)
)
```

```
df['job_content']
```

```
!pip install neattext
```

# Text Cleaning

* Use neattext or basic Python functions to remove:
  * Emails
  * Special characters
  * Numbers
  * Lowercase

```
import neattext.functions as nfx

df['cleaned_text'] = df['job_content'].apply(nfx.remove_emails)
df['cleaned_text'] = df['cleaned_text'].apply(nfx.remove_special_characters)
df['cleaned_text'] = df['cleaned_text'].apply(nfx.remove_numbers)
df['cleaned_text'] = df['cleaned_text'].str.lower()
```

```
df['cleaned_text']
```

# Train-Test Split

```
from sklearn.model_selection import train_test_split

X = df['cleaned_text']
y = df['fraudulent']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

# Feature Extraction

Here we use TF-IDF(Term Frequency-Inverse Document Frequency) , which is a type of feture extraction.

```
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)  # You can change this number
X_train_vec = tfidf.fit_transform(X_train).toarray()
X_test_vec = tfidf.transform(X_test).toarray()
```

# Apply Scaling

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_vec)
X_test_scaled = scaler.transform(X_test_vec)
```

# Apply SMOTE

Synthetic Minority Oversampling Technique (SMOTE) is a statistical technique for increasing the number of cases in your dataset in a balanced way.

```
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train_scaled, y_train)
```

#Convert Text to Sequences

```
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Parameters
max_words = 5000
max_len = 200

# Tokenizer
tokenizer = Tokenizer(num_words=max_words, oov_token='<OOV>')
tokenizer.fit_on_texts(X_train)

# Convert text to sequences
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Pad sequences
X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding='post')
```

#Build CNN Model

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout

cnn_model = Sequential([
    Embedding(input_dim=max_words, output_dim=128, input_length=max_len),
    Conv1D(128, 5, activation='relu'),
    GlobalMaxPooling1D(),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

cnn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history_cnn = cnn_model.fit(X_train_pad, y_train, epochs=5, batch_size=64, validation_split=0.2)
```

#Build RNN (LSTM) Model

```
from tensorflow.keras.layers import LSTM

rnn_model = Sequential([
    Embedding(input_dim=max_words, output_dim=128, input_length=max_len),
    LSTM(64, return_sequences=False),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

rnn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history_rnn = rnn_model.fit(X_train_pad, y_train, epochs=5, batch_size=64, validation_split=0.2)
```

#Evaluate CNN Models

```
from sklearn.metrics import classification_report

# CNN
y_pred_cnn = (cnn_model.predict(X_test_pad) > 0.5).astype(int)
print("CNN Classification Report:\n", classification_report(y_test, y_pred_cnn))
```

#Evaluate RNN Models

```
# RNN
y_pred_rnn = (rnn_model.predict(X_test_pad) > 0.5).astype(int)
print("RNN (LSTM) Classification Report:\n", classification_report(y_test, y_pred_rnn))
```

#Accuracy for CNN vs RNN

```
import matplotlib.pyplot as plt

plt.plot(history_cnn.history['val_accuracy'], label='CNN Val Acc')
plt.plot(history_rnn.history['val_accuracy'], label='RNN Val Acc')
plt.title("Validation Accuracy: CNN vs RNN")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

# Build Deep Neural Network (Keras)

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

dnn_model = Sequential()
dnn_model.add(Dense(256, input_dim=X_resampled.shape[1], activation='relu'))
dnn_model.add(Dropout(0.3))
dnn_model.add(Dense(128, activation='relu'))
dnn_model.add(Dropout(0.3))
dnn_model.add(Dense(1, activation='sigmoid'))

dnn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

dnn_model.fit(X_resampled, y_resampled, epochs=10, batch_size=64, validation_split=0.2)
```

# Evaluate the Model

```
y_pred_prob = dnn_model.predict(X_test_scaled)
y_pred = (y_pred_prob > 0.5).astype(int)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

#  Multilingual Fake Job Posting Prediction
 Multilingual job posting fraud detection system using:
- **DNN** model for English inputs
- **XLM-Roberta** model for other languages

```
import pandas as pd
import torch
from transformers import XLMRobertaTokenizer, XLMRobertaForSequenceClassification
from langdetect import detect
import torch.nn.functional as F
```

```
# Load XLM-Roberta for multilingual prediction
xlmr_tokenizer = XLMRobertaTokenizer.from_pretrained('xlm-roberta-base')
xlmr_model = XLMRobertaForSequenceClassification.from_pretrained('xlm-roberta-base', num_labels=2)  # 0: REAL, 1: FAKE
```

```
# Simulated DNN model for English (stubbed logic for demo)
def predict_with_dnn(text):
    # Demo logic: mark as FAKE if "earn money" or "work from home" is in text
    lower = text.lower()
    return 1 if any(phrase in lower for phrase in ["earn money", "work from home"]) else 0
```

```
# 🔁 Multilingual prediction function
def multilingual_predict(text):
    try:
        lang = detect(text)
    except:
        lang = 'unknown'

    if lang == 'en':
        pred = predict_with_dnn(text)
        model_used = 'DNN'
    else:
        inputs = xlmr_tokenizer(text, return_tensors="pt", truncation=True, padding=True)
        with torch.no_grad():
            outputs = xlmr_model(**inputs)
        logits = outputs.logits
        probs = F.softmax(logits, dim=1)
        pred = torch.argmax(probs).item()
        model_used = 'XLM-R'

    label = 'REAL' if pred == 0 else 'FAKE'
    return label, model_used
```

```
# 🧪 Sample multilingual job postings
test_jobs = [
    "Work from home and earn big money now!",  # English (scam-like)
    "Trabajo desde casa y gana mucho dinero!",  # Spanish
    "Nous recrutons un ingénieur logiciel expérimenté.",  # French
    "हम एक अनुभवी सॉफ़्टवेयर इंजीनियर की तलाश कर रहे हैं।",  # Hindi
]

# Run predictions
for job in test_jobs:
    label, model = multilingual_predict(job)
    print(f"📄 {job}\n→ Prediction: {label}   ({model} used)\n")
```
